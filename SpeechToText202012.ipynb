```
# -*- coding: utf-8 -*-
import json
from ibm_watson import SpeechToTextV1
from ibm_watson.websocket import RecognizeCallback, AudioSource
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator

#{API key}を自分のAPI鍵に置き換える。
authenticator = IAMAuthenticator('{API key}')
speech_to_text = SpeechToTextV1(
    authenticator=authenticator
)

#SSLの無効化を有効にする
speech_to_text.set_disable_ssl_verification(True)

#speech_to_text.set_service_url('{url}')の'{url}'を自分のurlに置き換える。
#Speech to textの管理ページへアクセスし、API鍵の下に表示されているurlを使います。
speech_to_text.set_service_url('{url}')

class MyRecognizeCallback(RecognizeCallback):
    def __init__(self):
        RecognizeCallback.__init__(self)

    def on_data(self, data):
        print(json.dumps(data, indent=1, ensure_ascii=False))

        # 変数を文字と一緒にprintする場合、f''で文を括る。変数は{}で括る。
        #print(f'transcript: {data["results"][0]["alternatives"][0]["transcript"]}')
        #print(f'confidence: {data["results"][0]["alternatives"][0]["confidence"]}')
        #
        #print(f'transcript: {data["results"][1]["alternatives"][0]["transcript"]}')
        #print(f'confidence: {data["results"][1]["alternatives"][0]["confidence"]}')
        #
        #print(f'transcript: {data["results"][2]["alternatives"][0]["transcript"]}')
        #print(f'confidence: {data["results"][2]["alternatives"][0]["confidence"]}')

        # ループ文で表示
        #data["results"]の個数を取得し、変数cntへ代入
        cnt = len(data["results"])
        for num in range(cnt): 
          #以下2行を修正すると全てのtranscript, confidenceを表示できる
          print(f'transcript: {data["results"][0]["alternatives"][0]["transcript"]}')
          print(f'confidence: {data["results"][0]["alternatives"][0]["confidence"]}')
    
    def on_error(self, error):
        print('Error received: {}'.format(error))

    def on_inactivity_timeout(self, error):
        print('Inactivity timeout: {}'.format(error))

myRecognizeCallback = MyRecognizeCallback()
#with open('join(dirname(__file__), './.', 'info-lady1-yomikomigakanryoushimashita1.mp3'),'rb') as audio_file:
# ファイルパス(audio_file)と、音声ファイルの拡張子(content_type)は自分でアップロードしたものに置き換えてください。
# 音声データが日本語以外の場合はmodelも適当な言語モデルに置き換える。 
audio_file = open("/content/drive/MyDrive/ML2020/rusuden_04.wav", "rb")
audio_source = AudioSource(audio_file)
speech_to_text.recognize_using_websocket(
        audio=audio_source,
        content_type='audio/wav',
        recognize_callback=myRecognizeCallback,
        model='ja-JP_BroadbandModel',
        #keywords=['生産', 'tornado', 'tornadoes'],
        #keywords_threshold=0.5,
        max_alternatives=0)
```
